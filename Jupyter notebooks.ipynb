## Cells
A python notebook is composed of cells. Cells can be of type: 
- Code cells: Input and output of live code that is run in the kernel.
- MarkDown cells: Narrative text with embedded LaTeX equations
- Heading cells: 6 levels of hierarchical organization and formatting
- Raw cells: Unformatted text that is included, without modification, when notebooks are converted to different formats using nbconvert.


```python
#This a code cell for a python kernel
a = 100
```

```python
#The memory is kept from one cell to another
print a * 2
```

Markdown cells allow you to write in raw manner in an intuitive way.

For a section just use #, for a subsection use ## (for example, I wrote ## Cells). The notebook will automatically convert it to a proper heading.

In markdown cells you also have LateX:

```python
import pandas as pd

dataframe = pd.DataFrame(X, columns=X_names)
dataframe["Y"] = Y
dataframe["Y_name"] = Y_names[0]
dataframe["Y_name"][dataframe["Y"] == 1] = Y_names[1]
dataframe
```

# More advance options
## Magic jupyter notebook commands

- Magic commands are called via % or %%
- Some can always be used, %matplotlib inline allows the notebook to display an image in the output
- Some allow timing of cells.
- Some can even be seen as more advance as they rely on python packages. %%R allows one to use R with an python kernel.


```python
## %matplotlib inline 
## you only have to put it once ...
from scipy.misc import imread
import matplotlib.pylab as plt

cute_cat_pic = imread("cat.jpeg")
plt.imshow(cute_cat_pic)
plt.show()
```

## Interactions with different languages
It is possible via the magic commands. Combining R and python in one document is possible.

A really cool statistical package is available in R but not in python? No worry, just install the python package r2py. And play around with IPython magic commands! (%R is for a one liner whereas %%R is for code blocks)


You will be able to pull/push data from R to python.

```python
%load_ext rpy2.ipython
```

```python
%R -i dataframe
```

```python
%%R
## we have just pushed data from python to R
print(dim(dataframe))
head(dataframe)
```

```python
%%R
### pushing data from R to python
p_value <- 0.0000001
```

```python
%R -o p_value
## p value will now exists in
## the python environnement.
```

```python
print p_value
```

## Remote kernels and cluster
My favourite feature are the remote kernels. You can (via some network handling/tunnelling) connect your notebook to a remote machine and still connect to it via your browser.

Pros:
- It would always be running. You can easily run long codes without worrying about putting your computing laptop in your bag.
- You can use priviliged hardware and ease the weight on your computer.
- You can have a very cheap computer or a very thin one with no computing power.
- You can directly plug it in a cluster (SGE cluster for instance) and run jobs directly from the notebook. (However this is an advance feature and it wouldn't be python doing the job submissions)

Cons:
- Be careful of evil IT turning off/rebooting the computer.
- It can be challinging setting up the tunneling, proxy settings and passing other security measures.
